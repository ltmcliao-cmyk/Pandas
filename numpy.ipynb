# 大數據下，兩條list對應index的值相乘

```python
"""
python寫法
"""
import time
import random
a = random.choices(range(10),k=1000000)
b = random.choices(range(10),k=1000000)

start_time = time.time()
c = []
for i in range(len(a)):
    c.append(a[i]*b[i])

print(f"run time at k=1000000 :{round((time.time()-start_time),6)}")

```

```python
"""
numpy寫法
"""
a_np = np.array(a)
b_np = np.array(b)
start_time = time.time()
c = a_np*b_np
print(f"run time at k=1000000 :{round((time.time()-start_time),6)}")
```

# Broadcast

```python
"""
不採用broadcast的list乘法
"""
import time
import random
a = random.choices(range(10),k=1000000)
b = np.full(1000000,2)

start_time = time.time()
c = a*b
print(f"run time at k=1000000 :{round((time.time()-start_time),6)}")

```

```python
"""
採用broadcast的list乘法
"""
a = random.choices(range(10),k=1000000)
b = 2

start_time = time.time()
c = a*b

print(f"run time at k=1000000 :{round((time.time()-start_time),6)}")

```

## 製造一百萬個2

```python
"""
製造1000000個2
"""
start_time = time.time()
# List Comprehension
a = np.array([2 for _ in range(1000000)])
a
print(f"run time at k=1000000 :{round((time.time()-start_time),6)}")



```

```python
"""
製造1000000個2
"""
start_time = time.time()

# List Multiplication
b = np.array([2] * 1000000)
b
print(f"run time at k=1000000 :{round((time.time()-start_time),6)}")

```

```python
"""
製造1000000個2
"""
start_time = time.time()
# np.ones
c = np.ones(1000000)*2
c
print(f"run time at k=1000000 :{round((time.time()-start_time),6)}")

```
